```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://scienti.minciencias.gov.co/gruplac/jsp/visualiza/visualizagr.jsp?nro=00000000008807'
```

```python
r = requests.get(url, verify=False)
soup = BeautifulSoup(r.text, 'lxml')
```

```python
#print(soup)
```

```python
NomGrupo = soup.find("span", class_='celdaEncabezado')
NombreGrupo = NomGrupo.text
NombreGrupo
```

```python
#Filtrar todas las tablas
table = soup.find_all('table')
df = pd.read_html(str(table))
#Numero de de tablas 
NumTablas = len(df)
NumTablas
```

```python
#Creacion de dataframe consolidado
dfconsolidado = pd.DataFrame()
dfconsolidado["Nombre del Grupo"] = [NombreGrupo]
```

```python
#CICLO 
dfcon = pd.DataFrame()
dfcon["Nombre del Grupo"] = [NombreGrupo]
#Inicio del ciclo
for i in range (0,10):
    #Tabla de datos 
    df = pd.read_html(str(table))[0]
    #Imputacion de los NAN
    df= df.fillna('Sin dato')
    #Tabla 1 
    tamaño = df.shape
    Nombre = df[0][0]
    Dato = []
    for j in range(1,tamaño[0]):
            Dato.append(df[0][j]+":  "+df[1][j]+"...   ")
    Datos = " ".join(Dato)
    Datos
    dfcon[Nombre]= [Datos]
    #Siguentes tablas
    df = pd.read_html(str(table))[i]
    tamaño = df.shape
    Nombre = df[0][0]
    Datos = []
    for j in range(1,tamaño[0]):
            Datos.append(df[0][j])
    Datos = " ".join(Datos)
    dfcon[Nombre]= [Datos] 
    
dfcon
```
